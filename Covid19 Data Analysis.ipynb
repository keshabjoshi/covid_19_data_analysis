```python
#import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset
dataset=pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#delete the unless columns
df=dataset.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
#aggregate the rows by the country
corona_dataset_aggregated=dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#Visalise data related to a country
corona_dataset_aggregated.loc["China"]
```

```python
# Ensure numeric data
corona_dataset_aggregated.fillna(0, inplace=True)
corona_dataset_aggregated = corona_dataset_aggregated.apply(pd.to_numeric, errors='coerce')

corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Nepal"].plot()
corona_dataset_aggregated.loc["Afghanistan"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
corona_dataset_aggregated.loc["China"].plot()
plt.legend()
```

```python
#calculate a good measure
corona_dataset_aggregated.loc["China"].plot()
```

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
#calculate the first derivate of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
# maximum infection rate
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Nepal"].diff().max()
```

```python
countries=list(corona_dataset_aggregated.index)
max_infection_rates=[]

for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rates"]=max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
#Create new dataframe
corona_data=pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```

```python
corona_data
```

```python
#importing the dataset 2
happiness_report=pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report
```

```python
# Drop Useless columns
unless_cols=["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.drop(unless_cols,axis=1,inplace=True)
happiness_report.head()
```

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report.head()
```

```python
# Join the dataset
corona_data.shape
```

```python
happiness_report.shape
```

```python
data=corona_data.join(happiness_report, how="inner")
data
```

```python
# Correlation matrix
correlation_matrix = data.corr()

# Display correlation matrix
correlation_matrix

# Visualize the correlations as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```
